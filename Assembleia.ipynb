```python

import requests
from bs4 import BeautifulSoup
import pandas as pd

pagina = requests.get("http://transparencia.al.al.leg.br/")
pagina = BeautifulSoup(pagina.content, 'html.parser')
nomes = []
nomes = pagina.findAll('td')

x = 0

listadaassembleia = []
while x < 100: #len(nomes):
    
    link = (nomes[x].find('a')['href'])
    print(nomes[x].text) ###Desmarque para acompanhar o processo
    
    paginainterna = requests.get(f"http://transparencia.al.al.leg.br/{link}")
    paginainterna = BeautifulSoup(paginainterna.content, 'html.parser')
    informacoes = paginainterna.findAll('td')
    dicionarioassembleia = {"Nome":informacoes[0].text, "Cargo":informacoes[1].text, "Remuneração Paradigma":informacoes[2].text, "Vantagens Pessoais":informacoes[3].text,"Subsídio diferença de Subsídio, função de confiança ou cargo comissionado":informacoes[4].text,"Indenizações":informacoes[5].text,"Vantagens Eventuais":informacoes[6].text,"Total de Créditos":informacoes[7].text,"Previdência Pública":informacoes[8].text,"Impostos de Renda":informacoes[9].text,"Descontos Diversos":informacoes[10].text,"Retenção por Teto Constitucional":informacoes[11].text,"Total de Débitos":informacoes[12].text,"Rendimento Líquido":informacoes[13].text,"Remuneração do Orgão de Origem":informacoes[14].text}
    listadaassembleia.append(dicionarioassembleia)
    
    x = x+1
dadoscompletos = pd.DataFrame(listadaassembleia)
display(dadoscompletos)
```
